```python
train_data = train.copy()
train = pd.get_dummies(train_data, columns = ['maincateg', 'platform'])
train.drop(['maincateg_Men', 'platform_Flipkart'], axis =1, inplace = True)
train.drop('title', axis=1, inplace=True)
train.drop('Offer %', axis=1, inplace = True)
train.head()
```

```python
print(test.shape[1])
print(test.isna().sum())
test.fillna(0,inplace =True)
```

```python
X = train.drop(columns=['price1'])
# y = np.log(train['price1'])
y = train['price1']
```

## Implement Pipeline for training and testing

Now we will implement a pipeline to streamline the training and testing process. first, we use a column transformer to encode categorical variables which is step one. After that, we create an object of our algorithm and pass both steps to the pipeline. using pipeline objects we predict the score on new data and display the accuracy.

```python
estimator = RandomForestRegressor()
param_grid = {
            "n_estimators"      : [40,100,200],
            "max_features"      : [ "sqrt", "log2"],
            "min_samples_split" : [1,2],
            "bootstrap": [True, False],
            "random_state" : [3,42],
            }

grid = GridSearchCV(estimator, param_grid, n_jobs=-1, cv=5)
grid.fit(X, y)
```

```python
print(grid.best_score_,grid.best_params_)
```

```python
estimator = RandomForestRegressor(n_jobs=-1).set_params(**grid.best_params_)
estimator.fit(X,y)
y_predict = estimator.predict(test)
print(y_predict)
```

## XGBRegressor with GridSearchCV

```python
from sklearn import preprocessing
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
```

```python
xgb1 = XGBRegressor()
parameters = {'nthread':[4], #when use hyperthread, xgboost may become slower
              'objective':['reg:linear'],
              'learning_rate': [.03, 0.05, .07], #so called `eta` value
              'max_depth': [5, 6, 7,8],
              'min_child_weight': [4],
              'silent': [1],
              'subsample': [0.7,0.8],
              'colsample_bytree': [0.7],
              'n_estimators': [500,1000]}

xgb_grid = GridSearchCV(xgb1,
                        parameters,
                        cv = 5,
                        n_jobs = 5,
                        verbose=True)

xgb_grid.fit(X,y)

print(xgb_grid.best_score_)
print(xgb_grid.best_params_)
```

```python
estimator2 = XGBRegressor(n_jobs=-1).set_params(**xgb_grid.best_params_)
estimator2.fit(X,y)
y_predict2 = estimator2.predict(test)
print(y_predict2)
```

```python
predictions_df2=pd.DataFrame(data={"id":test["id"],"price1":y_predict2})
predictions_df2.to_csv(path_or_buf="predictions_df2.csv",index=False,quoting=3,sep=',')
# np.savetxt(X=predictions_df, fname='price_pred.csv', fmt='%.0f')
```

```python
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV

params = {
    'num_leaves': [7, 14, 21, 28, 31, 50],
    'learning_rate': [0.1, 0.03, 0.003],
    'max_depth': [-1, 3, 5],
    'n_estimators': [50, 100, 200, 500],
}

grid = GridSearchCV(lgb.LGBMRegressor(random_state=3), params, scoring='r2', cv=5)
grid.fit(X, y)
print(grid.best_estimator_)
print(grid.best_params_)
```

```python
reg = lgb.LGBMRegressor(random_state=0)
reg.fit(X,y)

lgbm_tuned = grid.best_estimator_

r2_tuned = grid.best_params_
r2_regular = r2_score(y, reg.predict(X))
print(r2_regular)
```

```python
import lightgbm as lgb
lgb_model = lgb.LGBMRegressor(
    categorical_feature= [0,2,3,4,8,9,10],
    task = 'predict',
    application = 'regression',
    objective = 'root_mean_squared_error',
    boosting_type="gbdt",
    num_iterations = 2500,
    learning_rate = 0.05,
    num_leaves=15,
    tree_learner='feature',
    max_depth =10,
    min_data_in_leaf=7,
    bagging_fraction = 1,
    bagging_freq = 100,
    reg_sqrt='True',
    metric ='rmse',
    feature_fraction = 0.6,
    random_state=42)

lgb_model.fit(X,y) 


preds_lgb_model = lgb_model.predict(test)

parameters = {
    'task' : ['predict'],
    'boosting': ['gbdt' ],
    'objective': ['root_mean_squared_error'],
    'num_iterations': [  1500, 2000  ],
    'learning_rate':[  0.05, 0.005 ],
   'num_leaves':[ 7, 15, 31  ],
   'max_depth' :[ 10,15,25],
   'min_data_in_leaf':[15,25 ],
  'feature_fraction': [ 0.6, 0.8,  0.9],
    'bagging_fraction': [  0.6, 0.8 ],
    'bagging_freq': [   100, 200, 400  ],
     
}

gsearch_lgb = GridSearchCV(lgb_model, param_grid = parameters, n_jobs=6, verbose=10)
gsearch_lgb.fit(X,y)
 

print('best params')
print (gsearch_lgb.best_params_)
preds_lgb_model = gsearch_lgb.predict(test)
# rmse_lgb = np.sqrt(mean_squared_error(y_test, preds_lgb_model))
# print(" RMSE: %f" % (rmse_lgb ))
```
